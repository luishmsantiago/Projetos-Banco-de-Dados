```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.med.umich.edu/lrc/anatomy-tables/muscles_alpha.html'
soup = BeautifulSoup(requests.get(url).content, 'html.parser')
tables = soup.find_all('table')
```

```python
def generate_df(table):
    lines = []
    for i, row in enumerate(table.find_all('tr')[1:]):
        elem = 'th' if i == 0 else 'td'
        line = [row.text for row in row.find_all(elem)[:7]]
        lines.append(line)

    return pd.DataFrame(lines[1:], columns=lines[0])
```

```python
names = ['back_region', 'upper_limb', 'head_neck', 'thoracic', 'abdominal', 'pelvis_perenium', 'power_limb']
dfs = {name: generate_df(tables[i]) for i, name in enumerate(names, start=1)}
```
